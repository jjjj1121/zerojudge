```python
test = int(input())
for t in range(test):
    n = int(input())
    s=set()
    list1 =[]
    mmax =0
    for i in range(n):
        x = int(input())
        list1.append(x)
        if x not in s:
            s.add(x)
            if len(s)>mmax:
                mmax = len(s)
        else:
            for j in range(list1.index(x)):
                s.remove(list1[j])
            list1=list1[list1.index(x)+1:]
    print(mmax)
```
