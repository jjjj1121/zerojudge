```python
n = int(input())
max_list = [[-1*(2**31) for i in range(n//2)] for i in range(n//2)]
for i in range(n):
    row = list(map(int,input().split()))
    for j in range(n):
        max_list[i//2][j//2] =max(max_list[i//2][j//2],row[j])
for i in range(n//2):
    for j in range(n//2):
        print(f'{max_list[i][j]} ',end="")
    print()
```
