```python
n,m = map(int,input().split())
map_list = [[0 for i in range(m)] for j in range(n)]

for i in range(n):
    data = list(map(int,input().split()))
    for j in range(m):
        if data[j] == 1 :
            map_list[i][j]+=1
            if i-1>=0:
                map_list[i-1][j]+=1
            if j-1>=0:
                map_list[i][j-1]+=1
            if j+1<m:
                map_list[i][j+1]+=1
            if i+1<n:
                map_list[i+1][j]+=1
        elif data[j] == 2:
            for ii in range(n):
                map_list[ii][j]+=1
            for jj in range(m):
                map_list[i][jj]+=1
            map_list[i][j]-=1

for i in range(n):
    for j in range(m):
        print(f'{map_list[i][j]} ',end='')
    print()
```
