```python
n = int(input())
for i in range(n):
    x,y,z,w,n,m = map(int,input().split())
    carrots = list(map(int,input().split()))
    flag = 0
    count = 0
    if len(carrots) >0:
        for carrot in carrots:
            if flag == 1:
                hp = n*count
                m-=hp
                if m <0:
                    print("bye~Rabbit")
                    break
            if carrot == 1:
                m+=x
            elif carrot == 2:
                m+=y
            elif carrot == 3:
                m-=z
            elif carrot == 4:
                m-=w
                flag=1
                count+=1
            else:
                pass
            if m<=0:
                print("bye~Rabbit")
                break
        if m >0:
            print(f'{m}g')
    else:
        print(f'{m}g')
```
