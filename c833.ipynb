```python
from sys import stdin, stdout

n, m = map(int, stdin.readline().split())
rowCheck = [0 for i in range(n)]
columnCheck = 0

for i in range(n):
    block = stdin.readline()
    if "#" in block:
        rowCheck[i]=1
        block = block.replace('#', '1').replace('.', '0')
        columnCheck |= int(block, 2)
columnCheck=bin(columnCheck)
columnCheck = columnCheck[2:].zfill(m)
for i in range(n):
    if rowCheck[i]:
        stdout.write("#"*m+"\n")
    else:
        stdout.write(columnCheck.replace('1', '#').replace('0', 'X') + '\n')
```
